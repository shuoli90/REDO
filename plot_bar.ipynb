```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data extracted from the table for the requested methods under both scenarios
methods = ['IPA-GNN', 'E. IPS-GNN', 'REDO']
scenarios = ['Without Context', 'With Context']

acc = {
    'IPA-GNN': [68.3, 71.4],
    'E. IPS-GNN': [68.7, 71.6],
    'REDO': [75.1, 77.4]
}

wf1 = {
    'IPA-GNN': [64.8, 70.1],
    'E. IPS-GNN': [64.9, 70.9],
    'REDO': [69.9, 74.5]
}

x = np.arange(len(methods))  # the label locations
width = 0.2  # reduced the width of the bars

# Set font size
plt.rcParams.update({'font.size': 22})

fig, ax = plt.subplots(figsize=(8, 8))  # Square figure

# Plotting Acc
rects1 = ax.bar(x - width, [acc[m][0] for m in methods], width, label='Acc (Without Context)')
rects2 = ax.bar(x, [acc[m][1] for m in methods], width, label='Acc (With Context)')

# Plotting W. F1
rects3 = ax.bar(x + width, [wf1[m][0] for m in methods], width, label='W. F1 (Without Context)', alpha=0.7)
rects4 = ax.bar(x + 2 * width, [wf1[m][1] for m in methods], width, label='W. F1 (With Context)', alpha=0.7)

# Add labels and custom x-axis tick labels
ax.set_xlabel('Methods', fontsize=22)
ax.set_ylabel('Scores', fontsize=2)
ax.set_xticks(x + width / 2)
ax.set_xticklabels(methods)
ax.legend()

# Adjust y-axis limits to remove redundant space between 10 to 50
ax.set_ylim(60, 80)

# Increase tick label font size
ax.tick_params(axis='both', which='major', labelsize=22)

fig.tight_layout()
plt.grid()

plt.savefig('performance_comparison.pdf', dpi=300, bbox_inches='tight')
```
